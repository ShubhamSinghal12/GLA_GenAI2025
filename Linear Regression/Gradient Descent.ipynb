```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(X):
    return 2*X**2 - 4*X + 5
```

```python
X = np.linspace(-3,5,100)
y = f(X)
```

```python
plt.plot(X,y)
plt.show()
```

```python
def derF(X):
    return 4*X-4
```

```python
x = -2
alpha = 0.01
xv = [x]
yv = [f(x)]
```

```python
for i in range(200):
    x = x - alpha*derF(x)
    xv.append(x)
    yv.append(f(x))
```

```python
print(x)
```

```python
plt.plot(X,y)
plt.plot(xv,yv,'r-o')
plt.show()
```
