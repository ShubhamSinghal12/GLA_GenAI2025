```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv('./Training Data/Linear_X_Train.csv').values.reshape((-1,))
y = pd.read_csv('./Training Data/Linear_Y_Train.csv').values.reshape((-1,))
```

```python
X.shape
```

```python
y.shape
```

```python
plt.scatter(X,y)
plt.show()
```

```python
w1 = 50
w0 = 1
lr = 0.1
```

```python
yp = w1*X+w0
plt.scatter(X,y)
plt.plot(X,yp,c='r')
plt.show()
```

```python
w1 = w1 - lr*np.mean((yp-y)*X)
w0 = w0 - lr*np.mean(yp-y)
```

```python
print(w1,w0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
clf = LinearRegression()
```

```python
clf.fit(X.reshape((-1,1)),y)
```

```python
clf.intercept_
```

```python
clf.coef_
```

```python
r2_score(y,clf.predict(X.reshape((-1,1))))
```
