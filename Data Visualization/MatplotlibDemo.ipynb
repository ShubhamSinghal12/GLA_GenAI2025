````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
X = np.linspace(-2,2,1000)
y = np.sin(2*np.pi*X)

plt.figure(figsize=(10,2))
plt.plot(X,y)
plt.grid()
plt.title("Sin Wave")
plt.show()
````

````python
X = np.linspace(-2,2,1000)
y = np.cos(2*np.pi*X)

plt.figure(figsize=(10,2))
plt.plot(X,y)
plt.grid()
plt.title("Cos Wave")
plt.show()
````

````python
X = np.linspace(-2,2,1000)
sin = np.sin(2*np.pi*X)
cos = np.cos(2*np.pi*X)

plt.figure(figsize=(10,5))

plt.subplot(2,1,1)
plt.plot(X,sin)
plt.grid()
plt.title("Sin Wave")

plt.subplot(2,1,2)
plt.plot(X,cos)
plt.grid()
plt.title('Cos Wave')


plt.suptitle('Trignometric Functions')
plt.subplots_adjust(hspace=0.3)
plt.show()
````

````python
X = np.linspace(-3, 3 , 1000)

sin = np.sin(2*np.pi*X)
cos = np.cos(2*np.pi*X)
tan = np.tan(2*np.pi*X)
cot = 1/tan
sec = 1/cos
cosec = 1/sin

plt.figure(figsize=(10, 9))

plt.subplot(3, 2, 1)
plt.plot(X, sin)
plt.grid()
plt.title("Sin Wave")

plt.subplot(3, 2, 2)
plt.plot(X, cos)
plt.grid()
plt.title("Cos Wave")

plt.subplot(3, 2, 3)
plt.plot(X, tan)
plt.ylim(-5, 5)
plt.grid()
plt.title("Tan Wave")

plt.subplot(3, 2, 4)
plt.plot(X, cot)
plt.ylim(-5, 5)
plt.grid()
plt.title("Cot Wave")

plt.subplot(3, 2, 5)
plt.plot(X, cosec)
plt.ylim(-5, 5)
plt.grid()
plt.title("Cosec Wave")

plt.subplot(3, 2, 6)
plt.plot(X, sec)
plt.ylim(-5, 5)
plt.grid()
plt.title("Sec Wave")

plt.suptitle("Trignometric Functions")
plt.subplots_adjust(hspace=0.3)
plt.show()
````
