````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
x = [1,3]
y = [2,7]
plt.plot(x,y)
plt.show()
````

````python
X = np.linspace(-3,3,100)
y1 = 3*X+5
y2 = X**2+5

plt.plot(X,y1)
plt.plot(X,y2)

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
# plt.xlim((-4,4))
# plt.ylim((2.5,12))
plt.grid()
plt.show()
````

````python
x = np.arange(1,11)
y1 = [10,5,15,20,13,16,27,18,23,30]
y2 = 2*x+3
plt.plot(x,y1,'r',label = 'Price')
plt.plot(x,y2,'--o',label = 'Line')

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
plt.legend(loc = 'lower right')
plt.grid()
plt.show()
````

````python
plt.rcParams
````

````python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
x = np.arange(1,11)
y1 = 3*x+5
y2 = 2*x+3
plt.plot(x,y1,label = 'Price',color = 'r',marker = '*',markersize = 15,markerfacecolor = 'b')
plt.plot(x,y2,label = 'Line',color = 'g',marker = 'o',linestyle=':')

plt.xlabel('X',fontdict=font)
plt.ylabel('f(X) = Y',fontdict = font)
plt.title("Line Plot",fontdict = font)
plt.legend()
plt.grid()
# plt.savefig('LinePlot.jpg')
plt.show()
````

````python
plt.style.use('_classic_test_patch')
````

````python
X = np.linspace(-3,3,100)
y1 = 3*X+5
y2 = X**2+5

lines = plt.plot(X,y1,X,y2)

plt.setp(lines[0],color = 'r')
plt.setp(lines[1],color = 'g')

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
# plt.xlim((-4,4))
# plt.ylim((2.5,12))
plt.grid()
plt.show()
print(lines)
````

````python
X = np.linspace(-3,3,10)
y1 = 3*X+5
y2 = X**2+5

plt.scatter(X,y1,marker = '*',s = 30,c = 'r')
plt.scatter(X,y2,marker = '^',s = 40,c = 'cyan')

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
plt.grid()
plt.show()
````

````python
X = np.linspace(-2,2,1000)
y = np.sin(2*np.pi*X)

plt.figure(figsize=(10,2))
plt.plot(X,y)
plt.grid()
plt.title("Sin Wave")
plt.show()
````
